## Conversão de imagem RGB em imagem Grayscale

```python
import cv2 # importa a biblioteca do OpenCV
import numpy as np # Importa a biblioteca do NumPy, que analisa os complexos arrays da OpenCV
import matplotlib.pyplot as plt # importa a biblioteca matplotlib, que tem função de interações em imgs
#caso for usar o Google Colab com a OpenCV, usar a lib abaixo
from google.colab.patches import cv2_imshow
```

```python
#abrir a imagem
img = cv2.imread('tati.png',1)
#caso for usar o Google Colab com a OpenCV,
cv2_imshow(img)
```

```python
#mostrando a imagem colorida - caso você use Python no seu computador
#reutilize para exibir as imagens em outros códigos
#Voce pode descomentar o código apagando as aspas simples

# (desse jeito ta errado) cv2.imshow(img)
cv2_imshow(img)
'''
não é necessário
cv2.waitKey(0)
cv2.destroyAllWindows()
'''
```

```python
#aplicando conversão ponderada - converte imagem colorida para preto e branco
#img_grayscale_basic = 0.299*img[ : , : ,0] + 0.587*img[ : , : ,1] + 0.114*img[ : , : ,2]

#cv2
B, G, R = cv2.split(img) # separa a imagens para varias em canais de cor diferentes;
img_grayscale_pondered = 0.299*B+0.587*G+0.114*R # a cor que vai ficar a imagem

img_grayscale_pondered = np.array(img_grayscale_pondered, dtype=np.uint8) # esse datatype faz com que possua valores inteiros de 0-255 (max e min do RGB)

cv2_imshow(img_grayscale_pondered) # mostra a imagem ja convertida para o preto e branco
```

## Transformações

```python
#negativo

#img_negative[ : , : ,0] = 255 - img[ : , : ,0]
#img_negative[ : , : ,1] = 255 - img[ : , : ,1]
#img_negative[ : , : ,2] = 255 - img[ : , : ,2]

#Mude a variavel colorida para 1 caso queira colorida, o para 0 caso queira em escala de cinza
colorida = 1
img_in = cv2.imread('tati.png',colorida) # abre a imagem colorida ou cinza

img_out = 255 - img_in # pra inverter uma imagem é preciso subtrair o valor antigo por 255 (novovalor = 255 - valorantigo)

cv2_imshow(img_in) # mostra a imagem antiga
cv2_imshow(img_out) # mostra a imagem invertida
```

```python
#contraste e brilho

img_in = cv2.imread('tati.png',0) #abertura da imagem em preto e branco

#altere os valores tanto de a quanto de b

#constantes para mudança de brilho e constrate (chamados também de "gain and bias parameters")
a = -1
b = 1

img_out = a*img_in + b # mudança do contraste e brilho, na operação em pixels onde a imagem original é multiplicada por uma constante e depois somada por outra constante

img_out = np.array(img_out, dtype = np.uint8) # datatype np.uint8 pra acessar os inteiros de 0 a 255

cv2_imshow(img_in) # mostra a imagem antiga
cv2_imshow(img_out) # mostra a imagem com filtro
```

## Filtro Espacial

```python
#suavização

#você pode ler sobre o conceito aqui: https://docs.opencv.org/4.x/d4/d13/tutorial_py_filtering.html

img_in = cv2.imread('tati.png',0) # abertura da imagem em preto e branco

kernel = np.ones((5,5),np.float32)/25 # essa linha é a criação do kernel utilizando uma matriz de 1(np.ones) 5x5 multiplicada por 1/25 e
                                      # transforma esses valores em números reais de precisão única(np.float32)

img_out_1 = cv2.filter2D(img_in,-1,kernel) # o cv2.filter2D convoluciona o kernel com a imagem, mesclando as duas.


cv2_imshow(img_in) # mostra a imagem antiga
cv2_imshow(img_out_1) # mostra a imagem alterada com o filtro de suavização

```
